```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

# reading the image
img = imread('dog.jpg')
plt.rcParams['figure.figsize'] = [8,16]
np.shape(img)
```

```python
plt.imshow(img)
```

```python
img
```

```python
x = np.mean(img,-1)
```

```python
x
```

```python
image = plt.imshow(x)
image.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U,S,Vt = np.linalg.svd(x,full_matrices=False)
```

```python
S  = np.diag(S)
```

```python
for r in (4,16,50,600,602):
  xapprox = U[:,:r]@ S[:r,:r]@ Vt[:r,:]
  im = plt.imshow(xapprox)
  im.set_cmap('gray')
  plt.axis('off')
  plt.show()
```

```python
plt.semilogy(S)
plt.rcParams['figure.figsize'] = [5,4]
plt.show()
```
